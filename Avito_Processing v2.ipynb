```python
import pandas as pd
import numpy as np
import glob

path =r'C:\WorkData\Avito\Avito_Daily' # use your path
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
datalist_ = [] #list with all data
for file_ in allFiles:
    print('opening', file_)
    df = pd.read_csv(file_,index_col=None, names=['Name', 'Price', 'Date', 'Link'], header=0)
    #df.drop_duplicates(subset='Link', keep="last", inplace = True)
    datalist_.append(df)
print(str(len(datalist_))+' files opened')
GPUframe = pd.concat(datalist_, axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)

#GPUframe['Price'] = GPUframe['Price']
GPUframe = GPUframe.set_index('Date')
GPUframe['Price'] = GPUframe.apply(lambda x: x[1][:-3], axis=1)
```

```python
GPUframe['Name'] = GPUframe['Name'].str.lower()
```

```python
GPUframe.Name = GPUframe.Name.str.replace('gtx', 'GTX ')
GPUframe.Name = GPUframe.Name.str.replace('rx', 'RX ')
```

```python
GPUframe['Price'] =  GPUframe['Price'].str.replace('\s+','')
GPUframe = GPUframe[GPUframe['Price'].apply(lambda x: str(x).isdigit())]
```

## Regex combination 1

```python


#103 unique

import re

ShortNameList = []
Exceptions = []

NvidiaPattern = r'([G*][T*][X*])(\s*)(\d*)(\s*)(t?i?)'
#AMDPattern = r'([R*][X*](\s*)(\d*))'
AMDPattern = r'([R*][X*](\s*)(\d*)(v?e?g?a?)(\s*)(\d{2}))'

for name in GPUnotFarm['Name']:
    if len(re.findall(NvidiaPattern, name)) != 0:
        ShortNameList.append(re.search(NvidiaPattern, name).group(1)+' '+re.search(NvidiaPattern, name).group(3)+re.search(NvidiaPattern, name).group(5))
    elif len(re.findall(AMDPattern, name)) != 0:
        ShortNameList.append(re.search(AMDPattern, name).group(0))
    else:
        ShortNameList.append('Not what I need')
        Exceptions.append(name)# regex exceptions
print(ShortNameList, len(ShortNameList))
```

## Regex combination 2
#### 1st try = 103 unique
- NvidiaPattern = r'([G*][T*][X*])(\s*)(\d*)(\s*)(t?i?)'
- AMDPattern = r'([R*][X*](\s*)(\d*)(v?e?g?a?)(\s*)(\d{2}))'

#### 2nd try = 88 unique
improving NvidiaPattern
- NvidiaPattern = r'([G*][T*][X*])(\s*)(\d*)(\s*)(t{1}i{1})*'
improving AMDPattern
- AMDPattern = r'([R*][X*](\s*)(\d*)((v{1}e{1}g{1}a{1})?)(\s*)(\d{2-3})(\D))'

```python
import re

ShortNameList = []
Exceptions = []

NvidiaPattern = r'([G*][T*][X*])(\s*)(\d*)(\s*)(t{1}i{1})*'
AMDPattern = r'([R*][X*](\s*)(\d{3}))'
AMDVegaPattern = r'([R*][X*](\s*)(vega)(\s*)(\d*))'


for name in GPUnotFarm['Name']:
    if len(re.findall(NvidiaPattern, name)) != 0:
        ShortNameList.append(re.search(NvidiaPattern, name).group(0))
    elif len(re.findall(AMDPattern, name)) != 0:
        ShortNameList.append(re.search(AMDPattern, name).group(0))
    elif len(re.findall(AMDVegaPattern, name)) != 0:
        ShortNameList.append(re.search(AMDVegaPattern, name).group(0))    
    else:
        ShortNameList.append('Not what I need')
        Exceptions.append(name)# regex exceptions
print(ShortNameList, len(ShortNameList))
```

```python
#Dealing with whitespaces

ShortNameListFixed = []

for a in ShortNameList:
    a = ''.join(a.split())
    ShortNameListFixed.append(a)
    
print(ShortNameListFixed,len(ShortNameListFixed))
```
